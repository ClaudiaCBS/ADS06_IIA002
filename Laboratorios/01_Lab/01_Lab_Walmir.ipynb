```python
import seaborn as sns
import matplotlib.pyplot as plt

# Carrega o dataset "penguins" do Seaborn
df = sns.load_dataset("penguins")

# Cria um joint plot com dispersão e histogramas, colorido por espécie
sns.jointplot(data=df, x="flipper_length_mm", y="bill_length_mm", hue="species")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Carrega o dataset
df = sns.load_dataset("mpg")

# Cria um scatter plot com linhas de regressão por grupo
sns.lmplot(data=df, x="horsepower", y="mpg", hue="origin", ci=None)
plt.title("Relação entre Potência e Consumo por País de Origem")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Carrega o dataset
df = sns.load_dataset("taxis")

# Cria um boxplot para comparar a distribuição de 'fare' por 'passengers' e 'payment'
sns.boxplot(data=df, x="passengers", y="fare", hue="payment")
plt.title("Distribuição do Valor da Corrida por Número de Passageiros e Tipo de Pagamento")
plt.xlabel("Número de Passageiros")
plt.ylabel("Valor da Corrida (USD)")
plt.legend(title="Tipo de Pagamento")
plt.show()
```
